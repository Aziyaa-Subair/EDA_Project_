**POPULATION DENSITY AND ATTACKS**

```python
max_value_pop = df['Population_density'].max()
print(max_value_pop)
min_value_pop = df['Population_density'].min()
print(min_value_pop)
```

```python
ranges = [(50, 2000), (2001, 4000), (4001, 6000), (6001, 8000), (8001, 9999)]
range_counts = []
range_labels = []
for min_value, max_value in ranges:
    count = ((df['Population_density'] >= min_value) & (df['Population_density'] <= max_value)).sum()
    range_counts.append(count)
    range_labels.append(f"{min_value}-{max_value}")
    print(f"Population density range {min_value}-{max_value}: {count}")

```

```python
plot_data = pd.DataFrame({
    'Population Density Range': range_labels,
    'Count': range_counts
})

plt.figure(figsize=(10, 6))
sns.barplot(x='Population Density Range', y='Count', data=plot_data, palette='viridis')
plt.title('Population Density Distribution')
plt.xlabel('Population Density Range')
plt.ylabel('Count')
plt.ylim(12000, 13500)  # Set y-axis limits to zoom in on the differences
plt.xticks(rotation=45)
plt.show()
```

**Most attack happens in the population density range 6001-8000 with 13191**

**ELEVATION**

```python
max_elevation = df['Elevation'].max()
print("Maximum Elevation = ",max_elevation)
min_elevation = df['Elevation'].min()
print("Minimum Elevation =",min_elevation)
```

```python
costal_threshold = 200
mountain_threshold = 1000
costal_attack = ((df['Elevation'] >=min_elevation) & (df['Elevation'] < costal_threshold)).sum()
print("Number of attacks in costal area = ",costal_attack)
mountain_attack = ((df['Elevation'] <= max_elevation ) & (df['Elevation'] > mountain_threshold)).sum()
print("Number of attacks in mountain region=",mountain_attack)
other_attacks = ((df['Elevation']>=costal_threshold)&(df['Elevation'] <=mountain_threshold )).sum()
print("Number of attacks in Plateau region=",other_attacks)
```

```python
regions = ['Coastal', 'Mountain', 'Plateau']
attacks = [costal_attack, mountain_attack, other_attacks]

# Create the pie chart
plt.figure(figsize=(7, 7))
plt.pie(attacks, labels=regions, autopct='%1.1f%%', startangle=140)

# Add title
plt.title('Distribution of Cyber Attacks by Region')

# Show the plot
plt.show()
```

**Highest rate of attack happens in mountain region**

```python
# Filter the data for bus stops and metro stations
filtered_df = df[df['Public_transport_station'].isin(['bus stop', 'metro station'])]

# Group by and get the count of attack types and IoT devices for each station
grouped = filtered_df.groupby(['Public_transport_station', 'Cyber_attack_type', 'Iot_device_type']).size()

# Find the most common attack and IoT device for each public transport station
most_common_attacks = grouped.groupby('Public_transport_station').idxmax()

# Print the most common attack and IoT device for each station
for station in most_common_attacks.index:
    # Access the MultiIndex tuple and unpack it correctly
    attack_type, device_type = most_common_attacks[station][1], most_common_attacks[station][2]
    print(f"For {station}: Most common attack is '{attack_type}' and IoT device is '{device_type}'")
```

```python

# Filter the data for bus stops and metro stations
filtered_df = df[df['Public_transport_station'].isin(['bus stop', 'metro station'])]

# Group by and get the count of attack types and IoT devices for each station
grouped = filtered_df.groupby(['Public_transport_station', 'Cyber_attack_type', 'Iot_device_type']).size().reset_index(name='Count')

# Find the most common attack and IoT device for each public transport station
most_common_attacks = grouped.loc[grouped.groupby('Public_transport_station')['Count'].idxmax()]

# Create a figure with subplots
fig, axes = plt.subplots(2, 2, figsize=(18, 12))
fig.suptitle('Analysis of Cyber Attacks and IoT Devices at Public Transport Stations', fontsize=16)

# Plot 1: Most Common Cyber Attack Types
sns.barplot(x='Public_transport_station', y='Count', hue='Cyber_attack_type', data=most_common_attacks, ax=axes[0, 0], palette='viridis')
axes[0, 0].set_title('Most Common Cyber Attack Types')
axes[0, 0].set_xlabel('Public Transport Station')
axes[0, 0].set_ylabel('Count')
axes[0, 0].legend(title='Cyber Attack Type', bbox_to_anchor=(1.05, 1), loc='upper left')

# Plot 2: Most Common IoT Device Types
sns.barplot(x='Public_transport_station', y='Count', hue='Iot_device_type', data=most_common_attacks, ax=axes[0, 1], palette='magma')
axes[0, 1].set_title('Most Common IoT Device Types')
axes[0, 1].set_xlabel('Public Transport Station')
axes[0, 1].set_ylabel('Count')
axes[0, 1].legend(title='IoT Device Type', bbox_to_anchor=(1.05, 1), loc='upper left')

# Plot 3: Distribution of IoT Devices for Most Common Attacks
for i, station in enumerate(most_common_attacks['Public_transport_station'].unique()):
    station_data = grouped[(grouped['Public_transport_station'] == station) & 
                    (grouped['Cyber_attack_type'] == most_common_attacks.loc[most_common_attacks['Public_transport_station'] == station, 'Cyber_attack_type'].values[0])]
    sns.barplot(x='Iot_device_type', y='Count', data=station_data, ax=axes[1, i], palette='coolwarm')
    axes[1, i].set_title(f'Distribution of IoT Devices for Most Common Attack at {station}')
    axes[1, i].set_xlabel('IoT Device Type')
    axes[1, i].set_ylabel('Count')
    axes[1, i].tick_params(axis='x', rotation=45)

# Adjust layout
plt.tight_layout(rect=[0, 0, 1, 0.96])  # Adjust layout to prevent overlap
plt.show()
```

**The most common attack at a bus stop is SQL injection, while at a metro station, it is session hijacking; in both locations, a common IoT device used is an environmental sensor.**

## INSIGHTS:

Mountain Regions Experience the Highest Cyber Attack Rates: Due to limited infrastructure and weaker cybersecurity measures, remote mountain areas are prime targets for cyber attacks.

City Boundaries Are the Most Targeted Locations: High connectivity and the presence of critical infrastructure make city boundaries hotspots for cyber attacks.

Population Density Drives Attack Frequency: Areas with a population density of 6,001–8,000 people/km² experience the highest frequency of cyber attacks, highlighting urban areas as key targets.

Air Quality Index (AQI) Correlates with Cyber Attacks: Higher AQI levels, often found in urban areas, are associated with increased cyber attack frequency, likely due to the concentration of IoT devices and infrastructure.

Critical Infrastructure Faces Targeted Attacks: Specific infrastructure types, such as bridges (malware), dams (man-in-the-middle attacks), and data centers (DDoS), are highly vulnerable to distinct attack types.

Public Transport Hubs Are Vulnerable: Bus stops and metro stations are frequent targets, with SQL injection and session hijacking being the most common attacks, often involving environmental sensors.

Zero-Day Exploits Are the Most Common Attack Type: Zero-day exploits dominate the attack landscape, emphasizing the need for proactive vulnerability management and timely security updates.

Ransomware attacks are the most common, targeting critical healthcare devices like IoT-connected ECGs and remote patient monitors. This poses a significant risk to patient safety and data privacy.


## SUGGESTIONS

 Invest in robust cybersecurity infrastructure for remote mountain areas, including secure communication networks and IoT device hardening.
 
 Deploy advanced intrusion detection systems (IDS) and intrusion prevention systems (IPS) to monitor and protect critical infrastructure at city boundaries.

 Prioritize securing IoT devices in high-AQI areas, such as air quality sensors and industrial IoT systems, with encryption and authentication mechanisms.

 Develop tailored cybersecurity strategies for critical infrastructure sectors (e.g., bridges, dams, data centers) to address their unique vulnerabilities.

Secure IoT Sensors: Strengthen the security of environmental sensors and IoT devices at public transport hubs with encryption and access controls.

 Encrypt sensitive patient data stored on IoT-connected healthcare devices to prevent unauthorized access during ransomware attacks.
